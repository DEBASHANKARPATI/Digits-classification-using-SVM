<center><h1>Digits classification SVM</h1></center>

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import KFold,GridSearchCV,train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits_train=pd.read_csv("train.csv")
```

<center><h1>Data preparation</h1></center>

```python
Y=digits_train.pop("label")
```

```python
scaller =StandardScaler()
```

```python
X=scaller.fit_transform(digits_train)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.8,random_state=100)
```

<center><h1>Model Building</h1></center>

```python
#checking whether a linear model will be better against non linear or not 
model=SVC(kernel='linear')
```

```python
model.fit(Xtrain,Ytrain)
```

```python
Y_pred=model.predict(Xtest)
```

```python
metrics.accuracy_score(y_true=Ytest,y_pred=Y_pred)
```

```python
model=SVC(kernel='rbf')
```

```python
model.fit(Xtrain,Ytrain)
```

```python
Y_pred=model.predict(Xtest)
```

```python
metrics.accuracy_score(y_true=Ytest,y_pred=Y_pred)
```

Observation:- Turns out the non linear model performs better in terms of accuracy of the model ,Hence we will go with the non linear radial basis kernel

<center><h1>Hyperparameter tuning</h1></center>

```python
folds=KFold(n_splits=5,shuffle=True,random_state=4)
```

```python
hyperparameter={'gamma':[1e-1,1e-2,1e-3,1e-4],
                'C':[0.1,1,10,100,1000]}
```

```python
model=SVC(kernel='rbf')
```

```python
Gridmodel=GridSearchCV(estimator=model,return_train_score=True,verbose=1,scoring='accuracy',cv=folds,param_grid=hyperparameter)
```

```python
Gridmodel.fit(Xtrain,Ytrain)
```

```python
Gridmodel.best_params_
```

Observation-: we got C=10 and 0.001 as the best parameter , thus going forward we will use those values to get a optimal classifier .

<center><h1>Building final model with optimal hyperparameters</h1></center>

```python
model=SVC(kernel='rbf',C=10,gamma=0.001)
```

```python
model.fit(Xtrain,Ytrain)
```

```python
Y_pred=model.predict(Xtest)
```

```python
metrics.accuracy_score(y_true=Ytest,y_pred=Y_pred)
```
